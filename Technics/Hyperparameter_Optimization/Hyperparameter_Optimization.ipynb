```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./dataset/Iris_dataset.csv")
data.head()
```

```python
x = data.drop(["Species", "Id"], axis=1)
x.sample(5)
```

```python
y = data["Species"]
y.sample(5)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

<br><br>
## Manualy change params

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
model_2 = SVC(C=1, kernel="poly")
model_2.fit(x_train, y_train)
model_2.score(x_test, y_test)
```

```python
model_3 = SVC(C=10, kernel="rbf")
model_3.fit(x_train, y_train)
model_3.score(x_test, y_test)
```

```python
model_4 = SVC(C=10, kernel="poly")
model_4.fit(x_train, y_train)
model_4.score(x_test, y_test)
```

<br><br>
## Aouto change params

### GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "C": [0.1, 1, 10],
    "kernel": ["rbf", "linear", "poly"]
}
```

```python
grid_search = GridSearchCV(estimator=model, param_grid=param_grid)
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.score(x_test, y_test)
```

<br><br>
### RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_dist = {
    "C": [0.1, 1, 10],
    "kernel": ["rbf", "linear", "poly"]
}
```

```python
randomized_search = RandomizedSearchCV(estimator=model, param_distributions=param_dist, n_iter=5)
randomized_search.fit(x_train, y_train)
```

```python
randomized_search.best_params_
```

```python
randomized_search.score(x_test, y_test)
```
